1) ## Struktura sloupce description

Sloupec `description` má datový typ **JSON**.

Obsahuje strukturovaná data o produktu ve formátu JSON, například:

{
  "brand": "Dell",
  "category": "laptop",
  "cpu": "Intel i7",
  "ram_gb": 16
}

JSON obsahuje různé vlastnosti produktu, například:
- category – kategorie produktu
- brand – značka
- type – typ produktu
- size, length, power apod.

Struktura není pro všechny produkty stejná, ale většinou obsahuje klíč `category`.

2) ## Vysvětlení dotazu

Dotaz vybírá:

- `name` – název produktu
- `category` – hodnotu z JSON atributu `description`

Funkce `JSON_VALUE(description, '$.category')` získá hodnotu klíče `category` z JSON objektu.

Výsledkem je seznam produktů a jejich kategorií.

```python
-- Vytáhnu si název a k tomu kategorii, kterou vypreparuju z toho JSONu
SELECT
  name,
  JSON_VALUE(description, '$.category') AS category
FROM product;
```

3) Dotaz zobrazí všechny produkty, které jsou uložené v určitém skladu.

```python
-- Chci vidět všechno, co máme v prvním skladu
SELECT *
FROM product
WHERE warehouse_id = 1; -- Tady si můžu ID změnit podle potřeby
```

4) Dotaz zobrazí všechny produkty, které mají v JSON sloupci description v atributu category hodnotu cable.

```python
-- Najde to všechny produkty, co mají v JSONu kategorii nastavenou na 'cable'
SELECT *
FROM product
WHERE JSON_VALUE(description, '$.category') = 'cable';
```

5) Dotaz zobrazí všechny produkty, které mají množství větší než 20. Výsledky jsou seřazené vzestupně podle počtu.

```python
-- Filtruju podle množství a pak to seřadím vzestupně
SELECT *
FROM product
WHERE quantity > 20
ORDER BY quantity ASC; -- Seřazeno od 21 kusů nahoru
```

6) Dotaz zobrazí všechny produkty, které byly uskladněny v březnu roku 2024.

```python
-- Vybere produkty s datem uskladnění od začátku do konce března
SELECT *
FROM product
WHERE stored_from BETWEEN '2024-03-01' AND '2024-03-31';
```
